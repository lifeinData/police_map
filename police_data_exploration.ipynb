### Graphing the data

```python
df = pd.read_csv('MCI_2014_to_2017.csv', delimiter='\t')
```

```python
df['occurrencedate'] = df['occurrencedate'].apply(lambda x : str(x).split('T0')[0])
df['occurrencedate'] = df['occurrencedate'].astype('datetime64[ns]')
df['Y'] = df['Y'].astype('str')
df['X'] = df['X'].astype('str')
```

```python
df_2014 = df[df['occurrencedate'].dt.year == 2014]
df_2015 =  df[df['occurrencedate'].dt.year == 2015]
df_2016 =  df[df['occurrencedate'].dt.year == 2016]
df_2017 =  df[df['occurrencedate'].dt.year == 2017]

df_l = [df_2014,df_2015,df_2016,df_2017]
```

```python
day_dict = {'Monday' : 1, 'Tuesday' : 2, 'Wednesday' : 3, 'Thursday' : 4, 'Friday' : 5, 'Saturday' : 6, 'Sunday' : 7}
```

```python
data = []
start_y = 2014
for df_year in df_l:

    df_grouped = df_year.groupby([ "occurrencedayofweek"]).size().to_frame(name = 'count').reset_index()
    df_grouped['day_id']  = df_grouped['occurrencedayofweek'].map(day_dict)
    df_grouped.sort_values('day_id', inplace = True)

    x = df_grouped['occurrencedayofweek']
    y = df_grouped['count']

    trace = go.Scatter (
        x = x,
        y = y,
        mode = 'lines+markers',
        name = start_y
    )
    
    data.append (trace)
    start_y = start_y + 1
    
layout = dict(title = '# of crimes commited throughout the week 2014-2017'.title(),
             xaxis = dict (title = 'Day of Week'),
             yaxis = dict(title = '# of crimes committed')
             )

fig = dict(data = data, layout = layout)
py.iplot(fig, filename = 'basic_line_count_day_2014')
```

```python

def bmap ():
    m = folium.Map(
        location=[43.662809, -79.390943],
        zoom_start=14,
        tiles='Stamen Toner'
    )
    
    return m

#df_subset = df[df['occurrencehour'] == 23]

def plot_markers(df):
    map_l = []
    for i in range (5):
        print (i)
        df_subset = df[df['occurrencehour'] == i]
        m = bmap()
        
        for a in df_subset['loc']:

            y = a.split()[0]
            x = a.split()[1]
            folium.features.Circle(
                radius = 8,
                location = [float(y), float(x)],
                color='#3186cc',
                fill=True
            ).add_to (m)

        map_l.append (m)
        print (len(map_l))
    return map_l

map_l= plot_markers(df)

@interact(x=4)
def maps(x):
    return map_l[x]
```
